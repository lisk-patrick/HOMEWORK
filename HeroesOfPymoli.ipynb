```python
import pandas as pd
import json
import numpy as np

file = "purchase_data.json"
hop_df = pd.read_json(file)
```

```python
plyr_count = pd.DataFrame({"Total Players": [hop_df["SN"].count()]})
plyr_count.head()
```

```python
item_count = hop_df["Item Name"].nunique()
avg_price = hop_df["Price"].mean()
total_purchases = hop_df["Item Name"].count()
total_sales = hop_df["Price"].sum()

purchasing_analysis = pd.DataFrame({"Unique Items":[item_count],"Average Price":[avg_price],"Number of Purchases":[total_purchases],
                                   "Total Revenue": [total_sales]})
purchasing_analysis.head()
```

```python
# creates a list of all male purchasing data
male_count = hop_df["Gender"].value_counts()["Male"]
male_revenue = hop_df.loc[hop_df["Gender"]== "Male",:]["Price"].sum()
male_avg_purchase = male_revenue / male_count
male_percent = male_count / total_purchases * 100
male_std = hop_df.loc[hop_df["Gender"]== "Male",:]["Price"].std()
male_normal = (hop_df.loc[hop_df["Gender"]=="Male",:]["Price"] - hop_df.loc[hop_df["Gender"]=="Male",:]["Price"].min() ) / (hop_df.loc[hop_df["Gender"]=="Male",:]["Price"].max() - hop_df.loc[hop_df["Gender"]=="Male",:]["Price"].min() )
male_normal_total = male_normal.sum()

male_data = [male_count, male_avg_purchase, male_revenue,  male_normal_total]
```

```python
# creates a list of all female purchasing data
female_count = hop_df["Gender"].value_counts()["Female"]
female_revenue = hop_df.loc[hop_df["Gender"]=="Female",:]["Price"].sum()
female_avg_purchase = female_revenue / female_count
female_percent = female_count / total_purchases * 100
female_std = hop_df.loc[hop_df["Gender"]=="Female",:]["Price"].std()
#female_normal = (hop_df.loc[hop_df["Gender"]=="Female",:]["Price"] - hop_df.loc[hop_df["Gender"]=="Female",:]["Price"].min() ) / (hop_df.loc[hop_df["Gender"]=="Female",:]["Price"].max() - hop_df.loc[hop_df["Gender"]=="Female",:]["Price"].min() )
fem_norm = male_normal = (hop_df.loc[hop_df["Gender"]=="Female",:]["Price"] - hop_df.loc[hop_df["Gender"]=="Female",:]["Price"].min() ) / (hop_df.loc[hop_df["Gender"]=="Female",:]["Price"].max() - hop_df.loc[hop_df["Gender"]=="Female",:]["Price"].min() )
female_normal_total = fem_norm.sum()

female_data = [female_count, female_avg_purchase, female_revenue,  female_normal_total]
```

```python
# creates a list of all other/non-disclosed purchasing data
other_count = hop_df["Gender"].value_counts()["Other / Non-Disclosed"]
other_revenue = hop_df.loc[hop_df["Gender"]=="Other / Non-Disclosed",:]["Price"].sum()
other_avg_purchase = other_revenue / other_count
other_percent = other_count / total_purchases * 100
other_std = hop_df.loc[hop_df["Gender"]=="Other / Non-Disclosed",:]["Price"].std()
other_normal = (hop_df.loc[hop_df["Gender"]=="Other / Non-Disclosed",:]["Price"] - hop_df.loc[hop_df["Gender"]=="Other / Non-Disclosed",:]["Price"].min() ) / (hop_df.loc[hop_df["Gender"]=="Other / Non-Disclosed",:]["Price"].max() - hop_df.loc[hop_df["Gender"]=="Other / Non-Disclosed",:]["Price"].min() )
other_normal_total = other_normal.sum()
other_data = [other_count, other_avg_purchase, other_revenue,  other_normal_total]
```

```python
gender_df = pd.DataFrame({"Gender":["Male", "Female", "Other/Non-Disclosed"],
                                    "Percentage of Players": [male_percent, female_percent, other_percent],
                                   "Total Count": [male_count,female_count,other_count]})
gender_demographics = gender_df.set_index("Gender")
gender_demographics.head()
```
