```python
#  Analysis:
# - Two treatments are shown to reduce the tumor volume and maintain a low number of metastatic sites.
#      All other treatments follow the trend of the placebo, proving they are not effective.
#     Capomulin shows a reduction of 19.47% and Ramicane shows a 22.32% reduction in tumor size.
# - The average growth of tumor without treatment can be shown from the placebo tumor growth (51% growth)
#       Ketapril (57% growth) actually accelerated the tumor growth rate as well as Stelasyn (52%) and 
#         Naftisol (53%).
# - The number of metastatic sites shows the number of secondary tumors. No treatment was succesful
#         in preventing these sites, but Ramicane was the most effective, showing only 1.25 metastatic
#         sites over the 45 day trial period. Capomulin was second with 1.5 metastatic sites.
```
